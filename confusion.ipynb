```python
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
array = [[48  , 610],
[7  , 44142]]
ar = np.array(array)
df_cm = pd.DataFrame(ar,index=['positive / fraud', 'negative / normal'], columns = ['positive / fraud', 'negative / fraud'])

plt.figure(figsize = (10,7))
gif = plt.gcf()
sn.heatmap(df_cm, annot = True, fmt = '' ,annot_kws={"size": 18}).set_title("accuracy = 98.6229830160466  , recall / fraud detection = 87.27272727272727 ")
gif.savefig("modified_bayesian.png")
```
